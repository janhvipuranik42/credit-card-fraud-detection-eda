```python
# Credit Card Fraud Detection - Exploratory Data Analysis (EDA)

'''This project performs Exploratory Data Analysis on a credit card transaction dataset to understand patterns between fraudulent and legitimate transactions.'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Class'].value_counts()
```

```python
fraud = df['Class'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(fraud, labels=['Normal','Fraud'], autopct='%1.2f%%')
plt.title("Fraud vs Normal Transaction Percentage")
plt.show()
```

```python
fraud_df = df[df['Class'] == 1]
fraud_df.head()
```

```python
sns.countplot(x='Class', data=df)
plt.title("Fraud vs Normal Transactions")
plt.xlabel("Class (0 = Normal, 1 = Fraud)")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(fraud_df['Amount'], bins=30)
plt.title("Fraud Transaction Amount Distribution")
plt.show()
```

```python
sns.boxplot(x='Class', y='Amount', data=df)
plt.title("Fraud vs Normal Transaction Amount")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.scatter(df['Time'], df['Amount'], alpha=0.5)
plt.title("Time vs Transaction Amount")
plt.xlabel("Time")
plt.ylabel("Amount")
plt.show()
```

```python
## Key Insights
'''
1. Fraud transactions are extremely rare compared to normal transactions.
2. Most transactions have small amounts.
3. Fraud transactions show different patterns compared to legitimate ones.
4. Data visualization helps in understanding potential fraud patterns.'''
```
